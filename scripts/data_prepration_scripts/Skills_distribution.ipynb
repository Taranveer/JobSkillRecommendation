```python
data_path = "../../skills/Skill_List_count.txt"

skills_dict = dict()
with open(data_path) as f:
    line = f.readlines()
    for skill_cnt in line:
        skill_cnt =  skill_cnt.rstrip()
        try:
            skill, count = skill_cnt.split(" : ")
            count = int(count)
        except:
            pass
        print skill, count
        skills_dict[skill] = count
```

```python
skills_freq_counts = np.array(skill_cnts, dtype=np.float32)
percentage_dist = skills_freq_counts/np.sum(skills_freq_counts)
percentage_dist
```

```python
cummulative_dist = []
running_sum = 0
for x in percentage_dist:
    running_sum+=x
    cummulative_dist.append(running_sum)
```

```python
fig, ax = plt.subplots()
# the size of A4 paper
fig.set_size_inches(18, 8)

sns.tsplot(cummulative_dist , range(len(cummulative_dist)))
plt.show()
```
